# 11\. Unstructured Search (Grover's Algorithm)

### **1. Algorithm Profile**

  * **Name**: Groverâ€™s Search Algorithm
  * **Origin**: Discovered by **Lov Grover** in **1996** at Bell Labs.
  * **Type**: Oracular Algorithm (Amplitude Amplification).
  * **Speedup Class**: **Polynomial (Quadratic)**. It reduces the complexity from $O(N)$ (classical) to $O(\sqrt{N})$ (quantum).

### **2. Introduction & Context**

  * **The "Elevator Pitch"**: Imagine you have a phone book with $N$ names, but it's completely unsorted. To find a specific person's number, you'd have to check each entry one by one. Classically, you'd need to check roughly half the book on average. Grover's algorithm uses quantum mechanics to perform this search much faster. Instead of checking entries sequentially, it "rotates" the entire database toward the correct answer, allowing you to find the needle in the haystack with only $\approx \sqrt{N}$ checks.
  * **Real-World Use Cases**:
      * **Database Search**: Searching unsorted databases or lists.
      * **Constraint Satisfaction**: Solving logic puzzles (Sudoku, SAT problems) by searching the space of all possible solutions.
      * **Optimization**: Finding the minimum value in a complex dataset (Durr-Hoyer algorithm).
      * **Cryptanalysis**: Brute-forcing symmetric encryption keys (e.g., AES).
  * **Current Status**: It is a standard primitive in all quantum libraries. While simple in theory, implementing the "oracle" for real-world problems (like SAT) can require deep circuits.

### **3. Deep Theoretical Dive ðŸ§ **

#### **The Problem Statement**

Given a search space of size $N$ (where $N=2^n$) and an oracle function $f(x)$ such that:
$$f(x) = \begin{cases} 1 & \text{if } x \text{ is the winner } (w) \\ 0 & \text{otherwise} \end{cases}$$
Find the winning state $w$.

#### **The Classical Approach**

Check inputs $x$ at random or sequentially.

  * **Average Case**: $N/2$ queries.
  * **Worst Case**: $N$ queries.

#### **The Quantum Mechanism: Amplitude Amplification**

The algorithm works by repeatedly applying a "Grover Iterator" ($G$) that amplifies the amplitude of the winning state $|w\rangle$.

**1. Initialization**: Start with a uniform superposition of all states:
$$|s\rangle = H^{\otimes n}|0\rangle = \frac{1}{\sqrt{N}} \sum_{x=0}^{N-1} |x\rangle$$
This state $|s\rangle$ has a small overlap with the winner $|w\rangle$ ($\sin \theta = 1/\sqrt{N}$).

**2. The Grover Iterator ($G$)**:
Each step consists of two reflections:

  * **Oracle Reflection ($U_f$)**: Flips the phase of the winning state $|w\rangle$.
    $$U_f |x\rangle = (-1)^{f(x)} |x\rangle$$
    Geometrically, this reflects the state vector about the "non-winning" axis.
  * **Diffuser Reflection ($U_s$)**: Flips the state about the mean (the initial state $|s\rangle$).
    $$U_s = 2|s\rangle\langle s| - I$$

**3. Rotation**:
The combined operator $G = U_s U_f$ rotates the state vector by an angle of $2\theta$ towards the winning state $|w\rangle$.
After $k$ iterations, the state is:
$$|\psi_k\rangle = \sin((2k+1)\theta) |w\rangle + \cos((2k+1)\theta) |s^\perp\rangle$$

**4. Stopping Condition**:
We repeat this rotation until the amplitude of $|w\rangle$ is close to 1 (i.e., $\sin \approx 1$). This happens at $k \approx \frac{\pi}{4}\sqrt{N}$.

### **4. Practical Advantage & Complexity**

  * **Quantum Complexity**: $O(\sqrt{N})$ queries.
  * **Classical Complexity**: $O(N)$ queries.
  * **Resource Requirements**: $n$ qubits for the search space + ancilla for the oracle.
  * **Caveats**:
      * **Strict Quadratic limit**: It is proven that you cannot do better than $\sqrt{N}$ for unstructured search.
      * **Oracle Cost**: The "oracle" isn't free. If checking a solution takes a long time classically, it takes a long time quantumly too.
      * **Overshooting**: If you run the iterator too many times, you rotate *past* the solution and the probability decreases.

### **5. Implementation & Code Walkthrough (Qiskit) ðŸ’»**

We will implement Grover's Algorithm for **2 qubits** ($N=4$).

  * **Search Space**: $\{00, 01, 10, 11\}$
  * **Winner**: Let's arbitrarily choose $|11\rangle$ as the winner.

#### **Step 5.1: The Phase Oracle**

We need an oracle that flips the phase of the state $|11\rangle$ and leaves others unchanged.
For the state $|11\rangle$, the Controlled-Z ($CZ$) gate is the perfect phase flipper, as it applies a Z gate (phase flip) to the target if and only if the control is 1.

```python
import numpy as np
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
from qiskit.circuit.library import GroverOperator, MCMT, ZGate

# 1. Define the Oracle for state |11>
# A standard CZ gate flips the phase of |11>
def oracle_11():
    qc = QuantumCircuit(2)
    qc.cz(0, 1)
    qc.name = "Oracle (|11>)"
    return qc
```

#### **Step 5.2: The Diffuser Operator**

The diffuser reflects about the mean state $|s\rangle = |++\rangle$.
The standard construction is: $H^{\otimes n} \to X^{\otimes n} \to \text{Multi-Controlled Z} \to X^{\otimes n} \to H^{\otimes n}$.
For 2 qubits: $H \to X \to CZ \to X \to H$.

```python
# 2. Define the Diffuser for 2 qubits
def diffuser(n):
    qc = QuantumCircuit(n)
    # Apply H gates
    qc.h(range(n))
    # Apply X gates
    qc.x(range(n))
    
    # Multi-controlled Z (for 2 qubits, just CZ)
    # This flips the phase of the all-1 state (which corresponds to |00> before X gates)
    qc.cz(0, 1)
    
    # Uncompute X and H
    qc.x(range(n))
    qc.h(range(n))
    
    qc.name = "Diffuser"
    return qc
```

#### **Step 5.3: The Full Grover Circuit**

For $N=4$, the optimal number of iterations is exactly 1.
Calculation: $\frac{\pi}{4}\sqrt{4} = \frac{\pi}{2} \approx 1.57$. Rounding to nearest integer gives 1. (In fact, for $N=4$, 1 iteration gives 100% probability).

```python
n_qubits = 2
qc = QuantumCircuit(n_qubits)

# 1. Initialization (Uniform Superposition)
qc.h(range(n_qubits))

# 2. Grover Iterator (Oracle + Diffuser)
# Oracle
qc.append(oracle_11(), range(n_qubits))
# Diffuser
qc.append(diffuser(n_qubits), range(n_qubits))

# 3. Measurement
qc.measure_all()

print("Grover Circuit (Target |11>):")
qc.draw(output='text')
```

#### **Step 5.4: Execution & Analysis**

```python
# Run Simulation
sim = AerSimulator()
t_qc = transpile(qc, sim)
result = sim.run(t_qc, shots=1024).result()
counts = result.get_counts()

print("\nMeasurement Results:")
print(counts)
# plot_histogram(counts)
```

**Interpretation**:

  * You should see the state `11` with **100% probability**.
  * **Why 100%?** For the specific case of $N=4$, the rotation angle $\theta = 30^\circ$ ($\sin \theta = 1/2$).
  * One iteration rotates by $2\theta = 60^\circ$.
  * Total angle = $30^\circ + 60^\circ = 90^\circ$.
  * The state vector points exactly at the winner $|w\rangle$ ($\sin 90^\circ = 1$).
  * For larger $N$, the probability approaches 1 but may not reach exactly 100%.

# 11\. Unstructured Search (Grover's Algorithm)

### **1. Algorithm Profile**

  * **Name**: Groverâ€™s Search Algorithm
  * **Origin**: Discovered by **Lov Grover** in **1996** at Bell Labs.
  * **Type**: Oracular Algorithm (Amplitude Amplification).
  * **Speedup Class**: **Polynomial (Quadratic)**. It reduces the query complexity from $O(N)$ (classical) to $O(\sqrt{N})$ (quantum).

### **2. Introduction & Context**

  * **The "Elevator Pitch"**: Imagine finding a specific item in a large, unsorted database. Classically, you'd have to check items one by one, inspecting half the list on average. Grover's algorithm is a quantum method that finds the item much faster. It doesn't check items sequentially; instead, it views the search as a geometric rotation. By repeatedly "reflecting" the state vector, it rotates the probability amplitude from "wrong answers" toward the "correct answer," finding the needle in the haystack in just $\sqrt{N}$ steps.
  * **Real-World Use Cases**:
      * **Constraint Satisfaction**: Solving logic puzzles like Sudoku or 3-SAT by searching the space of all possible variable assignments.
      * **Graph Problems**: Finding triangles in graphs or solving the Traveling Salesman Problem.
      * **Cryptanalysis**: Brute-forcing symmetric encryption keys (e.g., breaking AES-128 requires only $2^{64}$ quantum operations).
  * **Current Status**: The algorithm is a standard benchmark. It runs on NISQ devices for small instances (2-5 qubits) but requires significant depth for useful problem sizes.

### **3. Deep Theoretical Dive ðŸ§ **

#### **The Problem Statement**

Given a search space of size $N=2^n$ and an oracle function $f(x)$ that marks one or more "winning" states $w$:
$$f(x) = \begin{cases} 1 & \text{if } x = w \\ 0 & \text{otherwise} \end{cases}$$
Find a winning state $w$.

#### **The Classical Approach**

In an unstructured list, there is no shortcut. You must query the oracle for each item.

  * **Complexity**: $O(N)$ queries.

#### **The Quantum Mechanism: Two Reflections make a Rotation**

Grover's algorithm works by **Amplitude Amplification**. It iterates two main operations:

1.  **Initialization**: Create a uniform superposition of all states: $|s\rangle = H^{\otimes n}|0\rangle$.
2.  **The Grover Iterator ($G$)**:
      * **The Oracle ($U_w$)**: Flips the phase of the winner(s).
        $$|x\rangle \xrightarrow{U_w} (-1)^{f(x)}|x\rangle$$
        Geometrically, this is a reflection about the axis orthogonal to the winner $|w\rangle$.
      * **The Diffuser ($U_s$)**: Flips the state about the mean (the initial state $|s\rangle$).
        $$U_s = 2|s\rangle\langle s| - I$$
        This is implemented as $H^{\otimes n} \to X^{\otimes n} \to \text{Multi-Controlled Z} \to X^{\otimes n} \to H^{\otimes n}$.

**Geometric Interpretation**: The state vector starts at $|s\rangle$. Each application of the Oracle and Diffuser rotates the vector by a fixed angle $\theta$ towards the winning state $|w\rangle$. After $k \approx \frac{\pi}{4}\sqrt{N/M}$ rotations (where $M$ is the number of solutions), the state vector aligns with $|w\rangle$, and measurement succeeds with near-certainty.

### **4. Practical Advantage & Complexity**

  * **Quantum Complexity**: $O(\sqrt{N/M})$ queries.
  * **Classical Complexity**: $O(N/M)$ queries.
  * **Resource Requirements**: $n$ qubits for the search space + ancilla.
  * **Caveats**: The quadratic speedup is sensitive to noise. If the number of solutions $M$ is unknown, a variation called **Quantum Counting** must be used first.

### **5. Implementation & Code Walkthrough (Qiskit) ðŸ’»**

We will implement two examples directly from your notebook: a standard 2-qubit search and a 3-qubit search with multiple solutions.

#### **Step 5.1: The 2-Qubit "Hello World" (Target $|11\rangle$)**

For 2 qubits ($N=4$), the algorithm finds the solution in exactly 1 iteration.

  * **Oracle**: To mark $|11\rangle$, we use a **Controlled-Z (CZ)** gate. It flips the phase only when both qubits are 1.

<!-- end list -->

```python
import numpy as np
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram

# 1. Initialize 2-qubit circuit
n = 2
qc = QuantumCircuit(n)

# 2. State Preparation: Uniform Superposition
qc.h([0, 1])

# 3. The Oracle (Marks |11>)
qc.cz(0, 1) 

# 4. The Diffuser (Reflection about mean)
qc.h([0, 1])
qc.z([0, 1]) # Apply Z to flip phase (part of reflection)
qc.cz(0, 1)
qc.h([0, 1])

# 5. Measurement
qc.measure_all()

print("2-Qubit Grover Circuit:")
qc.draw(output='text')
```

#### **Step 5.2: 3-Qubit Search (Multiple Solutions)**

Here we scale up to 3 qubits ($N=8$). We will use the specific oracle from your notebook which marks **two** solutions: $|101\rangle$ and $|110\rangle$.

  * **Oracle Logic**:
      * `qc.cz(0, 2)` flips phase for $|1x1\rangle$ (indices 5, 7).
      * `qc.cz(1, 2)` flips phase for $|x11\rangle$ (indices 3, 7).
      * State $|111\rangle$ (7) gets flipped twice ($(-1) \times (-1) = 1$), so it is **unmarked**.
      * State $|101\rangle$ (5) gets flipped once (Marked).
      * State $|110\rangle$ (6)? Wait, standard Qiskit ordering is $q_2, q_1, q_0$.
          * If state is $|110\rangle$ (decimal 6, $q_2=1, q_1=1, q_0=0$): `cz(0,2)` no trigger. `cz(1,2)` triggers? No, index 1 and 2.
          * Let's trace the notebook's specific logic: `qc.cz(0, 2)` and `qc.cz(1, 2)`.
          * This creates constructive interference for specific states. We will implement the Diffuser for general $n$ qubits.

<!-- end list -->

```python
def diffuser(n):
    """General Diffuser for n qubits"""
    qc = QuantumCircuit(n)
    # Apply H-gates
    qc.h(range(n))
    # Apply X-gates
    qc.x(range(n))
    
    # Multi-controlled Z (MCT followed by Z, or similar)
    # Equivalent to Multi-Controlled Toffoli where target is Z-rotated
    qc.h(n-1)
    qc.mct(list(range(n-1)), n-1) 
    qc.h(n-1)
    
    # Uncompute X and H
    qc.x(range(n))
    qc.h(range(n))
    return qc.to_gate(label="Diffuser")

# --- 3-Qubit Circuit Construction ---
n = 3
qc3 = QuantumCircuit(n)

# 1. Initialization
qc3.h(range(n))

# 2. Oracle from Notebook (Marks specific states)
# Marks states where (q0 & q2) OR (q1 & q2) is true? 
# Actually, acts as phase flip if specific conditions met.
qc3.cz(0, 2)
qc3.cz(1, 2)

# 3. Diffuser
qc3.append(diffuser(n), range(n))

# 4. Measure
qc3.measure_all()

print("\n3-Qubit Grover Circuit:")
qc3.draw(output='text')
```

#### **Step 5.3: Execution & Analysis**

```python
# Run Simulation
sim = AerSimulator()
t_qc3 = transpile(qc3, sim)
result = sim.run(t_qc3, shots=2048).result()
counts = result.get_counts()

# Visualize
print("\nMeasurement Results (3-Qubit):")
# We expect peaks at the marked solutions
print(counts)
plot_histogram(counts)
```

**Interpretation**:

  * You should see high probabilities for the states marked by the oracle.
  * Based on the oracle logic (`cz(0,2)` and `cz(1,2)`), the states that receive a net phase of -1 are the winners.
  * This example demonstrates Grover's ability to handle **multiple solutions** simultaneously. With 2 solutions in a space of 8, the optimal number of iterations is small (often just 1 is sufficient to boost probability significantly).

# 12\. Deutsch-Jozsa Algorithm

### **1. Algorithm Profile**

  * **Name**: Deutsch-Jozsa Algorithm.
  * **Origin**: Proposed by **David Deutsch** and **Richard Jozsa** in **1992**. It extended the earlier Deutsch Algorithm (1985).
  * **Type**: Oracular Algorithm (Query Complexity).
  * **Speedup Class**: **Exponential** (vs. Deterministic Classical). It requires 1 quantum query versus $2^{n-1}+1$ classical queries. Note: Probabilistically, it is only a constant speedup.

### **2. Introduction & Context**

  * **The "Elevator Pitch"**: Imagine you are given a black box (oracle) that takes in $n$ bits and outputs 0 or 1. You are promised that the box is either **Constant** (always outputs 0 or always 1) or **Balanced** (outputs 0 for half the inputs and 1 for the other half). Classically, to be 100% sure which type it is, you'd have to check more than half the possible inputs. The Deutsch-Jozsa algorithm can tell you the answer with 100% certainty using just **one** single check.
  * **Real-World Use Cases**:
      * **Testing Oracles**: It serves as a primitive for checking if a function is constant or has variation.
      * **Pedagogical Foundation**: It introduces the key concepts of **Phase Kickback** and **Interference** used in almost all advanced algorithms.
  * **Current Status**: It is a standard "Hello World" example in quantum libraries like Qiskit, often run on real hardware to benchmark coherence.

### **3. Deep Theoretical Dive ðŸ§ **

#### **The Problem Statement**

Given a function $f: \{0,1\}^n \to \{0,1\}$ with the promise that $f$ is either **Constant** or **Balanced**, determine which property holds.

#### **The Classical Approach**

  * **Deterministic**: In the worst case, you might check $2^{n-1}$ inputs and get the same answer (e.g., all 0s). You still don't know if it's constant or balanced until you check the $(2^{n-1} + 1)$-th input.
  * **Probabilistic**: If you check just a few random inputs and they are the same, it's exponentially likely to be constant.

#### **The Quantum Mechanism: Phase Kickback & Interference**

The algorithm uses a single "Hadamard sandwich" to solve the problem.

1.  **Initialization**: Prepare two registers: an $n$-qubit input $|0\rangle^{\otimes n}$ and a 1-qubit ancilla in the state $|-\rangle = \frac{|0\rangle - |1\rangle}{\sqrt{2}}$.
2.  **Superposition**: Apply Hadamard gates to the input register to create a uniform superposition $\sum |x\rangle$.
3.  **The Oracle (Phase Kickback)**: Apply the oracle $U_f$. Because the target is in $|-\rangle$, the value $f(x)$ is kicked back into the phase:
    $$U_f (|x\rangle |-\rangle) = (-1)^{f(x)} |x\rangle |-\rangle$$
    The state becomes:
    $$\frac{1}{\sqrt{2^n}} \sum_{x=0}^{2^n-1} (-1)^{f(x)} |x\rangle$$
4.  **Interference**: Apply Hadamard gates to the input register again. This performs a Boolean Fourier Transform.
      * The amplitude of the all-zero state $|0\dots0\rangle$ in the final superposition is:
        $$A_{0\dots0} = \frac{1}{2^n} \sum_{x=0}^{2^n-1} (-1)^{f(x)}$$
      * If $f$ is **Constant**: $f(x)$ is always 0 (or 1). The terms sum constructively. $|A_{0\dots0}| = 1$.
      * If $f$ is **Balanced**: Exactly half the terms are $+1$ and half are $-1$. They perfectly cancel out. $A_{0\dots0} = 0$.
5.  **Measurement**: Measure the input register.
      * If output is `00...0` $\to$ **Constant**.
      * If output is anything else $\to$ **Balanced**.

### **4. Practical Advantage & Complexity**

  * **Quantum Complexity**: **1 query**.
  * **Classical Complexity**: $O(2^n)$ queries (deterministic).
  * **Resource Requirements**: $n+1$ qubits. Circuit depth is constant (Hadamard-Oracle-Hadamard).
  * **Caveats**: The speedup disappears if we allow the classical computer to have a small probability of error (BPP vs BQP).

-----

### **5. Implementation & Code Walkthrough (Qiskit) ðŸ’»**

We will implement the Deutsch-Jozsa algorithm for a **3-qubit function** ($n=3$). We will create both a Constant Oracle and a Balanced Oracle and verifying the algorithm distinguishes them.

#### **Step 5.1: Creating the Oracle**

  * **Constant Oracle**: Does nothing (or flips output if constant 1).
  * **Balanced Oracle**: We can create a balanced function by using CNOTs. For example, $f(x) = x_0 \oplus x_1 \dots$ is balanced. We can also flip some output bits to make it more complex.

<!-- end list -->

```python
import numpy as np
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram

# 1. Define the Constant Oracle (Returns 0 or 1 for all inputs)
def constant_oracle(n):
    qc = QuantumCircuit(n+1)
    # Randomly decide if it's constant 0 or constant 1
    if np.random.randint(2) == 1:
        qc.x(n) # Flip output qubit to 1
    qc.name = "Constant Oracle"
    return qc

# 2. Define the Balanced Oracle (Returns 0 for half, 1 for half)
# Strategy: Apply CNOTs from input to output. This is parity function, which is balanced.
# We can also wrap qubits in X gates to change WHICH inputs give 1, preserving balance.
def balanced_oracle(n):
    qc = QuantumCircuit(n+1)
    b_str = "101" # Example string to wrap
    
    # Place X gates
    for qubit in range(len(b_str)):
        if b_str[qubit] == '1':
            qc.x(qubit)
            
    qc.barrier()
    
    # CNOTs to output (Function core)
    for qubit in range(n):
        qc.cx(qubit, n)
        
    qc.barrier()
    
    # Restore X gates
    for qubit in range(len(b_str)):
        if b_str[qubit] == '1':
            qc.x(qubit)
            
    qc.name = "Balanced Oracle"
    return qc
```

#### **Step 5.2: The Deutsch-Jozsa Circuit**

We wrap the oracle in the Hadamard sandwich.

```python
def dj_algorithm(oracle, n):
    qc = QuantumCircuit(n+1, n)
    
    # 1. Initialization
    # Input qubits to |+>
    for qubit in range(n):
        qc.h(qubit)
        
    # Output qubit to |->
    qc.x(n)
    qc.h(n)
    
    # 2. Apply Oracle
    qc.append(oracle, range(n+1))
    
    # 3. Interference (Hadamard again)
    for qubit in range(n):
        qc.h(qubit)
        
    # 4. Measurement (only measure input register)
    qc.measure(range(n), range(n))
    
    return qc

# Create circuits for both cases
n = 3
const_circuit = dj_algorithm(constant_oracle(n), n)
bal_circuit = dj_algorithm(balanced_oracle(n), n)

print("Balanced Circuit Diagram:")
bal_circuit.draw(output='text')
```

#### **Step 5.3: Execution & Analysis**

We expect the Constant circuit to output `000` with 100% probability, and the Balanced circuit to output anything *except* `000`.

```python
# Run Simulation
sim = AerSimulator()

# 1. Run Constant Case
t_const = transpile(const_circuit, sim)
res_const = sim.run(t_const, shots=1024).result().get_counts()

# 2. Run Balanced Case
t_bal = transpile(bal_circuit, sim)
res_bal = sim.run(t_bal, shots=1024).result().get_counts()

print(f"\nConstant Oracle Results: {res_const}")
print(f"Balanced Oracle Results: {res_bal}")

# Plot
# plot_histogram([res_const, res_bal], legend=['Constant', 'Balanced'])
```

**Interpretation**:

  * **Constant Case**: The result should be `{'000': 1024}`. All probability interferes constructively at the zero state.
  * **Balanced Case**: The result will be `{'111': 1024}` (or some mix of non-zero states). Since our oracle implemented a parity-like function, it often concentrates on specific non-zero states (like the all-ones state). Crucially, the count for `000` will be **0**.
  * **Conclusion**: A single query (one run of the circuit) gave us a bitstring. If it's all zeros, we know for sure the function is constant. If not, balanced.

# 13\. Bernstein-Vazirani Algorithm

### **1. Algorithm Profile**

  * **Name**: Bernstein-Vazirani Algorithm
  * **Origin**: Proposed by **Ethan Bernstein** and **Umesh Vazirani** in **1993**.
  * **Type**: Oracular Algorithm (Query Complexity).
  * **Speedup Class**: **Polynomial (Linear vs. Constant)**. It finds an $n$-bit secret string in **1 query**, whereas the best classical algorithm requires **$n$ queries**.

### **2. Introduction & Context**

  * **The "Elevator Pitch"**: Imagine a black box that holds a secret code (a string of bits). You can send it a query number, and it returns a special calculated response. If you want to find the secret code classically, you have to probe the box bit-by-bit, asking "Is the first bit 1?", "Is the second bit 1?", and so on. The Bernstein-Vazirani algorithm allows a quantum computer to send a single, complex superposition query that extracts the *entire* secret code at once.
  * **Real-World Use Cases**:
      * **Circuit Testing**: It is often used as a benchmark for calibrating quantum hardware because it requires creating a superposition of all states and interfering them perfectly.
      * **Learning Parity Functions**: It solves the problem of learning a linear boolean function from a single example.
  * **Current Status**: A standard "Hello World" algorithm. It runs successfully on current hardware (like IBM Quantum) for small $n$.

### **3. Deep Theoretical Dive ðŸ§ **

#### **The Problem Statement**

We are given an oracle function $f: \{0,1\}^n \to \{0,1\}$ defined by a secret string $s \in \{0,1\}^n$.
The function computes the bitwise inner product modulo 2:
$$f(x) = s \cdot x \pmod 2 = (s_0 x_0 \oplus s_1 x_1 \oplus \dots \oplus s_{n-1} x_{n-1})$$
**Goal**: Find the hidden string $s$.

#### **The Classical Approach**

To determine the $i$-th bit of $s$ (denoted $s_i$), we must query the function with an input $x$ that has a 1 only at position $i$ (e.g., $00\dots1\dots00$).

  * Query $x = 100\dots0 \implies f(x) = s_0$.
  * Query $x = 010\dots0 \implies f(x) = s_1$.
  * ...
    Total queries needed: **$n$**.

#### **The Quantum Mechanism: Fourier Sampling**

The algorithm uses the exact same "Hadamard Sandwich" circuit as Deutsch-Jozsa but interprets the result differently.

1.  **Initialization**: Prepare input register $|0\rangle^{\otimes n}$ and ancilla $|-\rangle$.
2.  **Superposition**: Apply $H^{\otimes n}$ to inputs. State: $\sum_{x} |x\rangle$.
3.  **Oracle Query (Phase Kickback)**: Apply $U_f$.
      * The value $f(x) = s \cdot x$ is kicked back into the phase.
      * State: $\frac{1}{\sqrt{2^n}} \sum_{x=0}^{2^n-1} (-1)^{s \cdot x} |x\rangle$.
4.  **Interference (Inverse QFT)**: Apply $H^{\otimes n}$ again.
      * This operation is the **Quantum Fourier Transform** over $\mathbb{Z}_2^n$.
      * The state $\sum (-1)^{s \cdot x} |x\rangle$ is exactly the Fourier transform of the state $|s\rangle$.
      * Therefore, applying $H^{\otimes n}$ (which is its own inverse) transforms the state directly into $|s\rangle$.
5.  **Measurement**: Measure the register. The result is the string $s$ with 100% probability.

### **4. Practical Advantage & Complexity**

  * **Quantum Complexity**: **1 query**.
  * **Classical Complexity**: **$n$ queries**.
  * **Resource Requirements**: $n+1$ qubits. Depth is $O(1)$ if parallel gates are allowed.

### **5. Implementation & Code Walkthrough (Qiskit) ðŸ’»**

We will implement the algorithm to find a **3-bit secret string**.

  * **Secret String**: $s = 011$ (Binary) $\implies s_2=0, s_1=1, s_0=1$.
  * **Oracle Function**: $f(x) = (0 \cdot x_2) \oplus (1 \cdot x_1) \oplus (1 \cdot x_0)$.

#### **Step 5.1: Constructing the Oracle**

The oracle $f(x) = s \cdot x$ can be implemented using CNOT gates.
If the $i$-th bit of $s$ is 1, we apply a CNOT between the $i$-th input qubit and the target qubit.

  * For $s=011$:
      * Bit 0 is 1 $\to$ CNOT(q0, target)
      * Bit 1 is 1 $\to$ CNOT(q1, target)
      * Bit 2 is 0 $\to$ No gate.

<!-- end list -->

```python
import numpy as np
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram

# 1. Define the Secret String
SECRET_STRING = '011' # s = 011
n = len(SECRET_STRING)

# 2. Create the Oracle Circuit
def bv_oracle(secret_s):
    n = len(secret_s)
    qc = QuantumCircuit(n+1)
    
    # The logic: f(x) = s . x
    # If s_i is 1, x_i contributes to the XOR sum.
    # We use CNOTs from input i to target (n) to perform XOR.
    
    # Note: String '011' -> s[0]='0' (Most Sig Bit), s[2]='1' (Least Sig Bit)
    # We need to match qubit ordering. 
    # Let's assume standard Qiskit order: q0 is LSB (rightmost char).
    # So s='011' means q2=0, q1=1, q0=1.
    
    reversed_s = secret_s[::-1] # Reverse to align index with string position
    
    for i, char in enumerate(reversed_s):
        if char == '1':
            qc.cx(i, n) # CNOT from input q_i to target
            
    qc.name = f"Oracle(s={secret_s})"
    return qc

# Visualize Oracle
oracle = bv_oracle(SECRET_STRING)
print(f"Oracle for s={SECRET_STRING}:")
oracle.draw(output='text')
```

#### **Step 5.2: The Full Algorithm Circuit**

We wrap the oracle in the Hadamard sandwich.

```python
# 3. Build the BV Circuit
qc = QuantumCircuit(n+1, n)

# A. Initialization
qc.h(range(n)) # Inputs to |+>
qc.x(n)        # Target to |1>
qc.h(n)        # Target to |-> (for Phase Kickback)

# B. Apply Oracle
qc.append(oracle, range(n+1))

# C. Interference (Inverse Fourier Transform)
qc.h(range(n))

# D. Measurement
qc.measure(range(n), range(n))

print("\nFull Bernstein-Vazirani Circuit:")
qc.draw(output='text')
```

#### **Step 5.3: Execution & Analysis**

We expect to measure `011` with 100% probability.

```python
# Run Simulation
sim = AerSimulator()
t_qc = transpile(qc, sim)
result = sim.run(t_qc, shots=1024).result()
counts = result.get_counts()

print(f"\nTarget Secret: {SECRET_STRING}")
print("Measurement Results:", counts)

# Verification
measured_string = max(counts, key=counts.get)
if measured_string == SECRET_STRING:
    print(">>> SUCCESS: Found the secret string in 1 shot!")
else:
    print(">>> FAILURE: Result does not match.")
```

**Interpretation**:

  * The result `011` appears 1024 times.
  * This confirms that the phase pattern created by the oracle $(-1)^{011 \cdot x}$ was perfectly transformed back into the computational basis state $|011\rangle$.
  * On a real noisy quantum computer, you might see small amounts of other strings (e.g., `010` or `001`) due to decoherence, but the correct answer will be the dominant peak.

# 14\. Simon's Algorithm

### **1. Algorithm Profile**

  * **Name**: Simonâ€™s Algorithm
  * **Origin**: Proposed by **Daniel Simon** in **1994**.
  * **Type**: Oracular Algorithm (Hidden Subgroup Problem).
  * **Speedup Class**: **Exponential**. It solves the problem in $O(n)$ queries, while any classical probabilistic algorithm requires $\Omega(2^{n/2})$ queries.

### **2. Introduction & Context**

  * **The "Elevator Pitch"**: Imagine a function that is "doubly periodic" in a strange way: for every input $x$, there is exactly one other input $y$ that gives the same output, and they are related by a secret "shift" $s$ (so $y = x \oplus s$). Classically, finding this secret shift is like finding a needle in a haystackâ€”you have to test pairs until you stumble upon a collision (the "Birthday Paradox" limit). Simon's algorithm uses quantum interference to find $s$ exponentially faster by filtering out all "wrong" information, leaving only equations that describe $s$.
  * **Real-World Use Cases**:
      * **Cryptanalysis**: Simon's algorithm inspired the attack on the Even-Mansour cipher and is relevant to the security of symmetric primitives (like Feistel networks) against quantum attackers.
      * **Theoretical Foundation**: It was the direct inspiration for Shor's Algorithm. Shor realized that if Simon could find a hidden period in the additive group ($\mathbb{Z}_2^n$), he could adapt it to find a period in the multiplicative group (factoring).
  * **Current Status**: A standard benchmark. It works well on NISQ devices because the oracle structure (CNOTs) is often simpler than modular exponentiation.

### **3. Deep Theoretical Dive ðŸ§ **

#### **The Problem Statement**

Given a function $f: \{0,1\}^n \to \{0,1\}^n$, we are promised that there exists a secret non-zero string $s \in \{0,1\}^n$ such that:
$$f(x) = f(y) \iff y = x \oplus s$$
**Goal**: Find the secret string $s$.

#### **The Classical Approach**

To find $s$, you need to find a "collision": two different inputs $x, y$ such that $f(x) = f(y)$. Then you can compute $s = x \oplus y$.

  * **Birthday Paradox**: If you check $k$ random inputs, you have roughly $k^2/2$ pairs. To have a high chance of a collision, you need $k \approx \sqrt{2^n} = 2^{n/2}$ queries.
  * **Complexity**: Exponential in $n$.

#### **The Quantum Mechanism: Constructive Interference on Orthogonality**

The algorithm looks almost identical to Bernstein-Vazirani but deals with a 2-to-1 function.

1.  **Superposition**: Prepare input $\sum |x\rangle$ and output $|0\rangle$.
2.  **Oracle Query**: Apply $U_f$. The state becomes:
    $$\sum_{x \in \{0,1\}^n} |x\rangle |f(x)\rangle$$
3.  **Measure Output (Conceptual)**: If we measure the output register and get a value $z = f(x_0)$, the input register collapses to the superposition of the two inputs that produce $z$:
    $$\frac{1}{\sqrt{2}} (|x_0\rangle + |x_0 \oplus s\rangle)$$
    *(Note: We don't actually need to measure the output; the math works the same due to deferred measurement principle).*
4.  **Interference (Hadamard)**: Apply $H^{\otimes n}$ to the input register.
    $$H^{\otimes n} \frac{1}{\sqrt{2}} (|x_0\rangle + |x_0 \oplus s\rangle) = \frac{1}{\sqrt{2^{n+1}}} \sum_{y} \left[(-1)^{x_0 \cdot y} + (-1)^{(x_0 \oplus s) \cdot y}\right] |y\rangle$$
    Looking at the phase term: $(-1)^{x_0 \cdot y} + (-1)^{x_0 \cdot y}(-1)^{s \cdot y} = (-1)^{x_0 \cdot y} (1 + (-1)^{s \cdot y})$.
      * If $s \cdot y = 1 \pmod 2$, the term is $(1 - 1) = 0$. **Destructive Interference**.
      * If $s \cdot y = 0 \pmod 2$, the term is $(1 + 1) = 2$. **Constructive Interference**.
5.  **Measurement**: We only measure values $y$ such that **$y \cdot s = 0 \pmod 2$**.

### **4. Practical Advantage & Complexity**

  * **Quantum Complexity**: $O(n)$ queries. Each measurement gives a linear equation involving $s$. With $\approx n$ equations, we can solve for $s$ using Gaussian elimination.
  * **Classical Complexity**: $O(2^{n/2})$.
  * **Resource Requirements**: $2n$ qubits (n input, n output).
  * **The Twist**: The algorithm doesn't give $s$ directly (unlike BV). It gives a system of equations. You must run the circuit $\approx n$ times to get enough info.

-----

### **5. Implementation & Code Walkthrough (Qiskit) ðŸ’»**

We will find a **3-bit secret string** $s = 110$ (binary).

  * **Constraint**: $f(x) = f(x \oplus 110)$.
  * **Oracle Strategy**: We need a function where output bits are linear combinations of inputs such that $f(x)$ ignores difference in the direction of $s$.
      * Ideally, we want $f(x)$ to act like a projection onto the subspace orthogonal to $s$.
      * For $s=110$, we need $n-1=2$ independent output bits.
      * $y_0 = x_2$ (since $s_2=0$, this bit passes through unchanged?) No, we need vectors $b$ where $b \cdot s = 0$.
      * Orthogonal basis for $s=110$: $v_1 = 001$, $v_2 = 110$? No, $110 \cdot 110 = 0 \pmod 2$.
      * Let's use a simple copy-based oracle logic: "Copy the bits that don't change when flipping by $s$".

#### **Step 5.1: Creating the Simon Oracle**

For $s=110$:

  * $x$ and $x \oplus 110$ must map to the same output.
  * $x \oplus 110$ flips bits 0 and 1 (in standard ordering) or bits 2 and 1 (in Qiskit ordering $q_2 q_1 q_0$). Let's assume $s$ is written as $s_2 s_1 s_0$.
  * If $s=110$, then $x$ and $x'$ differ in the first two bits.
  * A simple function satisfying this: $f(x) = x_0$ (the last bit) and $f(x) = x_0 \oplus x_1 \oplus x_2$ (parity)?
    Let's use the **Linear Implementation**:
    Find basis vectors $b_i$ orthogonal to $s$.
    For $s=110$ ($s \cdot y = 0$):

<!-- end list -->

1.  $y = 001$ (since $1 \cdot 0 = 0$).
2.  $y = 110$ (since $1\cdot 1 + 1\cdot 1 = 2 = 0$).
    The oracle will compute inner products with these basis vectors:

<!-- end list -->

  * Output 0: $x \cdot 001 = x_0$.
  * Output 1: $x \cdot 110 = x_2 \oplus x_1$.
    (We use a 3rd output bit just set to 0 or duplicate).

<!-- end list -->

```python
import numpy as np
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
from qiskit_aer.utils import insert_noise

# Target Secret
s = '110'
n = len(s)

def simon_oracle(s):
    # Create Quantum Circuit with 2n qubits (n input, n output)
    qc = QuantumCircuit(2*n)
    
    # Hardcoded linear mapping for s=110
    # We need outputs that are invariant under x -> x + s
    # Basis orthogonal to 110: [001], [110]
    
    # Output 0 = x . 001 = x[0] -> CNOT(0, n)
    # Output 1 = x . 110 = x[2] + x[1] -> CNOT(2, n+1), CNOT(1, n+1)
    
    # Note on ordering: String '110' usually means s2=1, s1=1, s0=0.
    # Orthogonal vectors: v such that v2*1 + v1*1 + v0*0 = 0.
    # v = 001 works (0+0+0=0).
    # v = 110 works (1+1+0=0).
    
    # Map x to output register
    
    # Copy bit 0 (x_0) to output bit 0
    qc.cx(0, n) 
    
    # Copy parity of x_2, x_1 to output bit 1
    qc.cx(2, n+1)
    qc.cx(1, n+1)
    
    # Output bit 2 can be redundant or 0. Let's leave it 0.
    # Or we can copy bit 0 again to ensure 2-to-1 mapping strictly?
    # Simon's requirement is f(x) = f(y) iff x = y + s.
    # Check: 
    # f(x) = (x0, x2+x1, 0)
    # f(x+s) = f(x + 110) = ( (x0+0), (x2+1 + x1+1), 0 ) 
    #        = (x0, x2+x1+2, 0) = (x0, x2+x1, 0).
    # It works! f(x) = f(x+s).
    
    qc.name = f"Oracle(s={s})"
    return qc
```

#### **Step 5.2: The Full Algorithm Circuit**

```python
# Setup Circuit
qc = QuantumCircuit(2*n, n) # Measure only n inputs

# 1. Initialize Inputs to |+>
qc.h(range(n))

# 2. Apply Oracle
qc.append(simon_oracle(s), range(2*n))

# 3. Interference (Hadamard on Inputs)
qc.h(range(n))

# 4. Measurement
qc.measure(range(n), range(n))

print("Simon's Algorithm Circuit:")
qc.draw(output='text')
```

#### **Step 5.3: Execution & Linear Solver Logic**

We run the circuit multiple times. Each time, we get a bitstring $y$.
We verify that $y \cdot s = 0 \pmod 2$.

```python
# Run Simulation
sim = AerSimulator()
t_qc = transpile(qc, sim)
result = sim.run(t_qc, shots=1024).result()
counts = result.get_counts()

print(f"\nTarget Secret s={s}")
print("Measurement Results (y values):")
sorted_counts = sorted(counts.items(), key=lambda x: x[1], reverse=True)

for bitstring, count in sorted_counts:
    # Check orthogonality
    # s = 110 (s2=1, s1=1, s0=0)
    # y = y2 y1 y0
    # dot = s2*y2 + s1*y1 + s0*y0
    
    # Convert binary string to list of ints
    y_val = [int(bit) for bit in bitstring] 
    # Note: bitstring 'y2 y1 y0' maps to indices 2, 1, 0 if printed standardly.
    # Let's assume standard big-endian printing.
    
    s_val = [int(bit) for bit in s]
    
    dot_product = sum([y * sv for y, sv in zip(y_val, s_val)]) % 2
    
    status = "Orthogonal (Good)" if dot_product == 0 else "FAILURE"
    print(f"Measured y={bitstring} | Count={count} | y.s = {dot_product} -> {status}")
```

**Interpretation**:

  * You should see outputs like `000`, `001`, `110`, `111`.
  * Check:
      * $y=001$: $1\cdot0 + 1\cdot0 + 0\cdot1 = 0$. (Orthogonal)
      * $y=110$: $1\cdot1 + 1\cdot1 + 0\cdot0 = 2 \equiv 0$. (Orthogonal)
  * **Solving**: If we get $y_1 = 001$ and $y_2 = 110$, we solve the system:
      * $s_0 = 0$
      * $s_2 + s_1 = 0 \implies s_2 = s_1$
      * Non-zero solution: $s_1=1 \implies s_2=1$. Result $s=110$.
  * **Conclusion**: The quantum computer filtered out all non-orthogonal strings, leaving us with a system of equations that uniquely identifies the secret period.

# 14\. Simon's Algorithm

### **1. Algorithm Profile**

  * **Name**: Simonâ€™s Algorithm
  * **Origin**: Proposed by **Daniel Simon** in **1994**.
  * **Type**: Oracular Algorithm (Hidden Subgroup Problem).
  * **Speedup Class**: **Exponential**. It solves the problem in $O(n)$ queries, whereas any classical probabilistic algorithm requires $\Omega(2^{n/2})$ queries.

### **2. Introduction & Context**

  * **The "Elevator Pitch"**: Simon's algorithm solves a "period finding" problem for a function that has a secret symmetry. Imagine a function where every output appears exactly twice. The two inputs that give the same output are related by a hidden "shift" string $b$ (i.e., $f(x) = f(y) \iff y = x \oplus b$). Classically, finding this shift is like finding a collision in a hash functionâ€”you have to check random inputs until you get lucky, which takes exponential time ($2^{n/2}$). Simon's algorithm uses quantum interference to filter out all incorrect answers, finding the hidden shift with a linear number of queries.
  * **Real-World Use Cases**:
      * **Theoretical Milestone**: This was the first algorithm to prove an **exponential** speedup over classical methods, directly inspiring Shor to discover his factoring algorithm.
      * **Cryptanalysis**: It provides a model for breaking symmetric cryptosystems (like Feistel networks) and authentication schemes (like Galois/Counter Mode) using quantum period finding.
  * **Current Status**: It is a standard benchmark for quantum hardware. Since the oracle uses only CNOT gates, it is more hardware-friendly than Shor's algorithm, allowing for demonstrations on current NISQ devices.

### **3. Deep Theoretical Dive ðŸ§ **

#### **The Problem Statement**

We are given a function $f: \{0,1\}^n \to \{0,1\}^n$. We are promised that there exists a secret non-zero string $b \in \{0,1\}^n$ such that:
$$f(x) = f(y) \iff y = x \oplus b$$
**Goal**: Find the secret string $b$.

#### **The Classical Approach**

The only way to find $b$ classically is to find a "collision"â€”two inputs $x$ and $y$ that produce the same output. Once found, $b = x \oplus y$.

  * **Birthday Paradox**: To have a 50% chance of finding a collision in a space of size $2^n$, you need to check approximately $\sqrt{2^n} = 2^{n/2}$ inputs.
  * **Complexity**: Exponential in $n$.

#### **The Quantum Mechanism: Orthogonal Sampling**

The algorithm uses interference to sample strings $z$ that reveal information about $b$.

1.  **Initialization**: Prepare two registers ($n$ qubits each) in state $|0\rangle|0\rangle$.
2.  **Superposition**: Apply Hadamard gates to the first register:
    $$\sum_{x} |x\rangle |0\rangle$$
3.  **Oracle Query**: Apply the function $f$. The state becomes:
    $$\sum_{x} |x\rangle |f(x)\rangle$$
4.  **Implicit Measurement**: If we measured the second register and got a value $y = f(x_0)$, the first register would collapse to the superposition of the two inputs that produce $y$:
    $$\frac{1}{\sqrt{2}} (|x_0\rangle + |x_0 \oplus b\rangle)$$
5.  **Interference**: Apply Hadamard gates again to the first register. This creates interference between the two states.
    $$H^{\otimes n} \frac{1}{\sqrt{2}} (|x_0\rangle + |x_0 \oplus b\rangle) = \sum_{z} \alpha_z |z\rangle$$
    The amplitude $\alpha_z$ for a string $z$ is non-zero **only if** $b \cdot z = 0 \pmod 2$.
      * If $b \cdot z = 1$, the terms destructively interfere ($1 - 1 = 0$).
      * If $b \cdot z = 0$, the terms constructively interfere ($1 + 1 = 2$).
6.  **Measurement**: Measuring the first register returns a string $z$ that is guaranteed to be orthogonal to the secret string $b$.

### **4. Practical Advantage & Complexity**

  * **Quantum Complexity**: $O(n)$ queries. Running the circuit $\approx n$ times gives us $n$ linear equations ($b \cdot z_i = 0$), which we can solve to find $b$.
  * **Classical Complexity**: $O(2^{n/2})$ queries.
  * **Resource Requirements**: $2n$ qubits ($n$ for input, $n$ for output).

-----

### **5. Implementation & Code Walkthrough (Qiskit) ðŸ’»**

We will implement the case for **3 qubits** with the secret string **$b = 110$**.
This means $b_2=1, b_1=1, b_0=0$.

#### **Step 5.1: The Simon Oracle**

To construct a function where $f(x) = f(x \oplus b)$, we can use a simple trick:

1.  Copy the input register to the output register (using CNOTs). This creates a 1-to-1 function $f(x)=x$.
2.  Identify the "least significant bit" of $b$ that is 1. Let's say it is index $k$.
3.  If the $k$-th input bit is 1, we XOR the output with $b$.
      * This ensures that $x$ and $x \oplus b$ map to the same output.
      * Example ($b=110$): The last non-zero bit is at index 1 ($b_1=1$).
      * If $x_1=0$, output $x$.
      * If $x_1=1$, output $x \oplus 110$.
      * Check: $000 \to 000$. $110 \to 110 \oplus 110 = 000$. (Collision created\!)

<!-- end list -->

```python
import numpy as np
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram

# 1. Define the Secret String
b = '110'
n = len(b)

def simon_oracle(b):
    """Returns a Simon oracle for the secret string b"""
    # Create a QuantumCircuit with 2n qubits
    qc = QuantumCircuit(2*n)
    
    # 1. Copy content of first register to second register (CNOTs)
    # This creates a 1-to-1 mapping initially
    for i in range(n):
        qc.cx(i, n+i)
        
    # 2. Introduce Collisions based on b
    # Find the least significant index 'k' where b[k] == '1'
    # Note: b string is '110' -> b[0]='1', b[1]='1', b[2]='0'
    # We reverse it to match Qiskit qubit ordering (q0 is rightmost char)
    b_reversed = b[::-1]
    
    pivot_index = -1
    for i, bit in enumerate(b_reversed):
        if bit == '1':
            pivot_index = i
            break
            
    if pivot_index != -1:
        # For every bit j where b[j] is 1, apply CNOT(pivot_index, output_j)
        for i, bit in enumerate(b_reversed):
            if bit == '1':
                # Control is the pivot qubit in input register
                # Target is the corresponding qubit in output register
                qc.cx(pivot_index, n + i)
                
    return qc

# Draw the Oracle
oracle = simon_oracle(b)
print(f"Oracle for b={b}:")
oracle.draw(output='text')
```

#### **Step 5.2: The Full Algorithm Circuit**

The standard "Hadamard Sandwich" applied to the input register.

```python
# Create the Algorithm Circuit
simon_circuit = QuantumCircuit(2*n, n)

# 1. Hadamard on Input Register
simon_circuit.h(range(n))
simon_circuit.barrier()

# 2. Apply Oracle
simon_circuit.append(oracle, range(2*n))
simon_circuit.barrier()

# 3. Hadamard on Input Register (Interference)
simon_circuit.h(range(n))

# 4. Measure Input Register
simon_circuit.measure(range(n), range(n))

print("Simon's Algorithm Circuit:")
simon_circuit.draw(output='text')
```

#### **Step 5.3: Execution & Verification**

We expect to measure strings $z$ such that $z \cdot b = 0 \pmod 2$.
For $b=110$: $1\cdot z_2 + 1\cdot z_1 + 0\cdot z_0 = 0$.
So $z_2$ must equal $z_1$.
Expected results: `000`, `001`, `110`, `111` (assuming $z_2=z_1$).

```python
# Run Simulation
sim = AerSimulator()
t_qc = transpile(simon_circuit, sim)
result = sim.run(t_qc, shots=1024).result()
counts = result.get_counts()

print(f"\nTarget Secret: {b}")
print("Measurement Results (z values):")

# Helper to verify dot product
def bdotz(b, z):
    accum = 0
    for i in range(len(b)):
        # Match string indices
        accum += int(b[i]) * int(z[i])
    return (accum % 2)

for z, count in counts.items():
    dot_product = bdotz(b, z)
    status = "âœ… Orthogonal" if dot_product == 0 else "âŒ Error"
    print(f"Measured: {z} | Count: {count} | b.z = {dot_product} ({status})")

plot_histogram(counts)
```

**Analysis**:

  * You will only see states where the dot product is 0.
  * **Post-Processing**: In a real scenario, you run this circuit $O(n)$ times to get a list of $z$ vectors. You then assume they are rows in a matrix and use **Gaussian Elimination** to solve for the non-zero vector $b$ that is orthogonal to all of them.
  * This transforms an exponential search problem into a polynomial linear algebra problem.

# 15\. Collision Finding

### **1. Algorithm Profile**

  * **Name**: BHT Algorithm (Brassard, HÃ¸yer, and Tapp)
  * **Origin**: Proposed by **Gilles Brassard**, **Peter HÃ¸yer**, and **Alain Tapp** in **1997**.
  * **Type**: Oracular Algorithm (Hybrid Search).
  * **Speedup Class**: **Polynomial**. It finds a collision in a 2-to-1 function in $O(N^{1/3})$ queries, improving on the classical $O(\sqrt{N})$ bound.

### **2. Introduction & Context**

  * **The "Elevator Pitch"**: The "Birthday Paradox" tells us that in a room of just 23 people, it's likely two share a birthday. This is a collision. Classically, finding a collision in a random list of size $N$ takes about $\sqrt{N}$ steps. The BHT algorithm beats this limit by combining classical memory with quantum search. It builds a small classical database of inputs and then uses **Grover's algorithm** to search the rest of the space for a match, finding a collision much faster than either method alone.
  * **Real-World Use Cases**:
      * **Cryptanalysis of Hash Functions**: Finding collisions in functions like SHA-2 or MD5 is a primary goal of attackers. While BHT applies to 2-to-1 functions, generalizations (like Ambainis's algorithm for Element Distinctness) apply to finding collisions in random functions.
  * **Current Status**: Theoretical primitive. The requirement for quantum random access to classical memory makes it hardware-intensive.

### **3. Deep Theoretical Dive ðŸ§ **

#### **The Problem Statement**

Given an oracle function $f: X \to Y$ where $|X|=N$, find two distinct inputs $x, y \in X$ such that $f(x) = f(y)$.
We assume the function is **2-to-1**, meaning every output in the range has exactly two pre-images.

#### **The Classical Approach: Birthday Attack**

  * Pick random inputs $x_1, x_2, \dots$ and store $(x_i, f(x_i))$ in a hash table.
  * Stop when you find a new $x_{new}$ such that $f(x_{new})$ is already in the table.
  * **Complexity**: $O(\sqrt{N})$ due to the Birthday Paradox.

#### **The Quantum Mechanism: Hybrid Search**

The BHT algorithm optimizes the trade-off between building a classical list and running a quantum search.

1.  **The Parameters**: Let $K = N^{1/3}$.
2.  **Classical Phase**:
      * Pick $K$ random inputs $\{x_1, \dots, x_K\}$.
      * Compute their values $f(x_i)$ and store the pairs $(x_i, f(x_i))$ in a classical lookup table (sorted or hashed for $O(1)$ access).
3.  **Quantum Phase**:
      * Construct a Grover Oracle $U_g$ that recognizes a collision with the classical list.
      * The oracle takes an input $|x\rangle$. It computes $f(x)$ and checks the classical table.
      * If $f(x)$ exists in the table (and $x$ is not the same pre-image we already stored), then $x$ is a **winner**.
      * Flip the phase if winner: $U_g |x\rangle = -|x\rangle$.
4.  **Grover Search**:
      * The search space is the remaining $N-K$ items.
      * Since the function is 2-to-1, each of the $K$ stored items has exactly one "partner" in the remaining set.
      * Thus, there are $K$ solutions in the search space of size $N$.
      * Grover's algorithm takes $O(\sqrt{N/K})$ steps to find one of these solutions.
5.  **Complexity Analysis**:
      * Classical part: $K$ queries.
      * Quantum part: $\sqrt{N/K}$ queries.
      * Balance them: $K = \sqrt{N/K} \implies K^{3/2} = N^{1/2} \implies K = N^{1/3}$.
      * **Total Complexity**: $O(N^{1/3})$.

### **4. Practical Advantage & Complexity**

  * **Quantum Complexity**: $O(N^{1/3})$.
  * **Classical Complexity**: $O(N^{1/2})$.
  * **Caveats**: Requires huge quantum-accessible classical memory (QRAM) to implement the oracle efficiently. Without efficient QRAM, the oracle query cost might increase, negating the speedup.

### **5. Implementation & Code Walkthrough (Qiskit) ðŸ’»**

We will simulate the BHT algorithm logic.

  * **Function**: A simple 2-to-1 mapping.
      * Domain: $0 \dots 7$ (3 qubits).
      * Range: $0 \dots 3$.
      * Mapping: $f(x) = x // 2$ (Integer division).
      * Pairs: $(0,1) \to 0$, $(2,3) \to 1$, $(4,5) \to 2$, $(6,7) \to 3$.
      * Collisions are $(0,1), (2,3), \dots$.
  * **Classical Part**: We pick a subset size $K=2$.
  * **Quantum Part**: Grover search for the "partners" of the classical subset.

#### **Step 5.1: Defining the Function and Classical Phase**

```python
import numpy as np
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
from qiskit.circuit.library import GroverOperator

# 1. Define the 2-to-1 Function
# f(x) = x // 2 (maps 0,1->0; 2,3->1; etc.)
def oracle_function(x):
    return x // 2

# 2. Classical Phase: Build the Table
N = 8
K = 2 # Number of classical queries
# Pick K random inputs. For demo, let's pick specific ones to ensure we don't just pick a pair immediately.
# Let's pick inputs 0 and 2.
classical_inputs = [0, 2]
lookup_table = {oracle_function(x): x for x in classical_inputs}

print(f"Classical Table: {lookup_table}")
# Table: {0: 0, 1: 2}
# This means we have seen output 0 (from input 0) and output 1 (from input 2).
# We are looking for their partners: 
# Partner of 0 is 1. (f(1)=0)
# Partner of 2 is 3. (f(3)=1)
```

#### **Step 5.2: The BHT Oracle**

The oracle checks if $f(x)$ is in `lookup_table`.
If yes, and $x$ is not the value already in the table, flip phase.

  * Target outputs: 0 and 1 are in table.
  * Inputs producing 0: {0, 1}. 0 is in table. **1 is the target**.
  * Inputs producing 1: {2, 3}. 2 is in table. **3 is the target**.
  * Inputs producing 2, 3: Not in table.
    **Winners**: $|1\rangle$ (001) and $|3\rangle$ (011).

We construct a Phase Oracle marking `001` and `011`.

```python
def bht_oracle():
    # We need to mark |1> (001) and |3> (011).
    # Logic: Mark if (q0=1) AND (q1=0 OR q1=1) AND (q2=0) ? 
    # Simplification:
    # |001> -> q2=0, q1=0, q0=1
    # |011> -> q2=0, q1=1, q0=1
    # Common features: q2=0, q0=1.
    # The value of q1 doesn't matter.
    # So we mark any state where q2=0 AND q0=1.
    
    qc = QuantumCircuit(3)
    
    # To mark state |0x1>:
    # Apply X to q2 (to make 0 trigger control)
    # Apply CZ between q2 and q0
    # Apply X to q2 (restore)
    
    qc.x(2)
    qc.cz(2, 0)
    qc.x(2)
    
    qc.name = "BHT Oracle"
    return qc

print("Oracle Circuit:")
bht_oracle().draw(output='text')
```

#### **Step 5.3: The Grover Search**

With $N=8$ and 2 solutions, 1 iteration is usually optimal.

```python
# 3. Setup Grover
n_qubits = 3
qc = QuantumCircuit(n_qubits)

# A. Superposition
qc.h(range(n_qubits))

# B. Grover Operator
oracle = bht_oracle()
grover_op = GroverOperator(oracle)
qc.append(grover_op, range(n_qubits))

# C. Measure
qc.measure_all()

# 4. Run
sim = AerSimulator()
t_qc = transpile(qc, sim)
result = sim.run(t_qc, shots=1024).result()
counts = result.get_counts()

print("\nMeasurement Results:")
print(counts)
```

#### **Step 5.4: Verification**

```python
print("\nChecking Collisions:")
for bitstring, count in counts.items():
    if count > 100:
        x_quantum = int(bitstring, 2)
        y_val = oracle_function(x_quantum)
        
        # Check against table
        if y_val in lookup_table:
            x_classical = lookup_table[y_val]
            if x_quantum != x_classical:
                print(f"FOUND COLLISION: Quantum({x_quantum}) and Classical({x_classical}) both map to {y_val}")
            else:
                print(f"Found existing element {x_quantum} (Not a new collision)")
        else:
            print(f"Quantum found {x_quantum} -> {y_val} (Not in table)")
```

**Interpretation**:

  * The quantum computer finds inputs `1` and `3` with high probability.
  * Comparing `1` to the classical table (`0`), we see $f(1)=0, f(0)=0$. **Collision found\!**
  * Comparing `3` to the classical table (`2`), we see $f(3)=1, f(2)=1$. **Collision found\!**
  * The algorithm successfully used classical "breadcrumbs" to direct the quantum search towards the colliding partners.

# 15\. Collision Finding

### **1. Algorithm Profile**

  * **Name**: The BHT Algorithm (Brassard, HÃ¸yer, and Tapp).
  * **Origin**: Proposed by **Gilles Brassard**, **Peter HÃ¸yer**, and **Alain Tapp** in **1997**.
  * **Type**: Oracular Algorithm (Hybrid Search).
  * **Speedup Class**: **Polynomial**. It finds a collision in a 2-to-1 function in **$O(N^{1/3})$** queries, beating the classical limit of $O(\sqrt{N})$ (the Birthday Paradox bound).

### **2. Introduction & Context**

  * **The "Elevator Pitch"**: The "Birthday Paradox" tells us that in a group of just 23 people, it's likely two share a birthday. This is finding a "collision" in a random function. Classically, finding a collision in a search space of size $N$ takes about $\sqrt{N}$ steps. The BHT algorithm beats this by using a team effort: a classical computer builds a small list of "seen" items, and then a quantum computer uses **Grover's search** to scan the remaining space for any item that matches one on the list. This hybrid approach is significantly faster than either machine working alone.
  * **Real-World Use Cases**:
      * **Cryptanalysis of Hash Functions**: Finding two different files that have the same hash (a collision) allows attackers to forge digital signatures. This algorithm provides the theoretical lower bound for the security of hash functions against quantum attacks.
      * **Breaking "Golden" cryptographic primitives**: Many proofs of security rely on the hardness of finding collisions.
  * **Current Status**: Theoretical. The algorithm requires **QRAM (Quantum Random Access Memory)** to efficiently query the classical list from within the quantum superposition, a hardware capability that is difficult to build.

### **3. Deep Theoretical Dive ðŸ§ **

#### **The Problem Statement**

Given a function $f: X \to Y$ where $|X|=N$. We are promised that $f$ is **2-to-1**, meaning every output $y$ has exactly two distinct inputs $x_1, x_2$ such that $f(x_1) = f(x_2) = y$.
**Goal**: Find any pair of distinct inputs $a, b$ such that $f(a) = f(b)$.

#### **The Classical Approach: The Birthday Attack**

  * **Method**: Pick random inputs and store them in a hash table. Check each new input to see if its output is already in the table.
  * **Analysis**: You need to check approximately $\sqrt{N}$ items to have a high probability of finding a match.
  * **Complexity**: $O(N^{1/2})$.

#### **The Quantum Mechanism: The Hybrid Trade-off**

The BHT algorithm balances two costs: the cost of building a classical list and the cost of searching for a match.

1.  **Classical Phase**:

      * Randomly select a subset of $K$ inputs: $S = \{x_1, \dots, x_K\}$.
      * Compute their values and store the pairs $(x_i, f(x_i))$ in a classical lookup table.
      * **Cost**: $K$ queries.

2.  **Quantum Phase**:

      * Define a Grover Oracle $U_g$ that recognizes a "collision" with the stored list $S$.
      * For any input $|x\rangle$, the oracle computes $f(x)$ and checks if $f(x)$ appears in the classical table (and $x \notin S$).
      * Since the function is 2-to-1, every $x_i$ in our list $S$ has exactly one "partner" $x'$ outside the list such that $f(x_i) = f(x')$.
      * Therefore, there are exactly **$K$ winning items** in the remaining search space of size $N$.
      * Use **Grover's Algorithm** to search for one of these $K$ winners.
      * **Cost**: $O(\sqrt{N/K})$ queries.

3.  **Optimization**:
    To minimize the total time $T = K + \sqrt{N/K}$, we set the two terms equal:
    $$K = \sqrt{\frac{N}{K}} \implies K^2 = \frac{N}{K} \implies K^3 = N \implies K = N^{1/3}$$

      * **Total Complexity**: $O(N^{1/3})$.

### **4. Practical Advantage & Complexity**

  * **Quantum Complexity**: $O(N^{1/3})$.
  * **Classical Complexity**: $O(N^{1/2})$.
  * **Resource Requirements**: $n$ qubits for the search, plus a classical memory of size $N^{1/3}$ that can be queried quantumly.
  * **Caveats**: The speedup depends entirely on the efficient implementation of the oracle "Check Table". This requires **QRAM** or a large quantum circuit to encode the table, which can be resource-heavy.

### **5. Implementation & Code Walkthrough (Qiskit) ðŸ’»**

We will simulate the BHT algorithm on a small scale.

  * **Function**: A 2-to-1 function on 3 qubits ($N=8$).
      * $f(x) = x \pmod 4$.
      * Pairs: $(0,4), (1,5), (2,6), (3,7)$.
      * i.e., $f(0)=0, f(4)=0$. Collision\!
  * **Classical Step**: Pick $K=2$ inputs. Let's say we pick $S=\{0, 1\}$.
      * Table: $\{f(0): 0, f(1): 1\} \to \{0: 0, 1: 1\}$.
  * **Quantum Step**: Search for inputs $x$ such that $f(x) \in \{0, 1\}$ but $x \notin \{0, 1\}$.
      * Targets: $4$ (matches 0) and $5$ (matches 1).

#### **Step 5.1: Creating the Oracle**

The oracle marks state $|x\rangle$ if $f(x)$ is in the lookup table.
For our example:

  * Table contains outputs $0, 1$.
  * $f(x) = x \pmod 4$ corresponds to the last 2 bits of $x$.
  * We mark $x$ if `last_2_bits(x)` is `00` (0) or `01` (1).
  * **Exclusion**: We must strictly ensure we don't find the items *already* in the list (0 and 1). But Grover's naturally finds "partners" if we initialize the superposition correctly or simply check $x \neq x_{stored}$ classically after measurement. For this simple oracle, we will mark all 4 items (0,1,4,5) and see what we get.

<!-- end list -->

```python
import numpy as np
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
from qiskit.circuit.library import GroverOperator

# 1. Define Problem Parameters
n_qubits = 3  # Domain size 8
# Function: f(x) = x % 4 (Last 2 bits)
# Stored outputs: {0, 1}
# Target inputs (Collision partners): 4 ('100'), 5 ('101')

def bht_oracle():
    # Mark states where f(x) is in {0, 1}.
    # f(x) is determined by the lower 2 qubits (q0, q1).
    # We want to mark states where the lower 2 bits are '00' or '01'.
    # This simplifies to: Mark if q1 is 0.
    # (00, 01 are valid; 10, 11 are not).
    
    qc = QuantumCircuit(n_qubits)
    
    # Phase Oracle Logic: Flip phase if q1 == 0.
    # We can do this by flipping q1, applying Z, flipping back.
    # Or simpler: Apply Z to q1? No, Z flips if q1=1.
    # We want flip if q1=0. So X -> Z -> X.
    
    qc.x(1)
    qc.z(1) # Applies -1 phase if q1 is 1 (which was originally 0)
    qc.x(1)
    
    qc.name = "Oracle (Match Table)"
    return qc

print("Oracle Circuit:")
bht_oracle().draw(output='text')
```

#### **Step 5.2: The Grover Search Loop**

We are searching a space of $N=8$.
Target items: $\{0, 1, 4, 5\}$ (all items mapping to stored outputs).
This is half the space ($M=4$).
Optimal iterations $\approx \frac{\pi}{4}\sqrt{8/4} \approx 1.1$.
Wait, if half the items are solutions, we don't need Grover's\! We can just sample.
*Correction*: BHT is useful when $K \ll N$. Here $K=2, N=8$ is dense.
Let's refine the target. Suppose we only stored value `0`.

  * Table: $\{f(0): 0\}$.
  * Target: Partner of 0 is `4`.
  * Solutions: $\{0, 4\}$. $M=2$.
  * Oracle: Mark if last 2 bits are `00`. (q0=0, q1=0).

<!-- end list -->

```python
# REVISED Oracle: Only search for collision with output 0.
# Target states: |000> (0) and |100> (4).
# Condition: q0=0 AND q1=0.

def bht_oracle_refined():
    qc = QuantumCircuit(3)
    # Mark states where q0=0 and q1=0.
    # Apply X to q0, q1. Then apply Multi-controlled Z (CCZ logic).
    # Then restore X.
    
    qc.x([0, 1])
    
    # Emulate CCZ using H and CCX (Toffoli)
    # Apply H to target (say q2), then CCX(q0, q1, q2)? 
    # No, CCZ acts on 3 qubits symmetrically.
    # Or simpler: We just need a phase flip if q0=1, q1=1 (after X gates).
    # We can use a CZ gate on q0, q1? No, that ignores q2.
    # If we only care about q0 and q1, a CZ on them works for ALL q2 values!
    # It marks |000> and |100> simultaneously.
    
    qc.cz(0, 1)
    
    qc.x([0, 1])
    qc.name = "Oracle (f(x)=0)"
    return qc
```

#### **Step 5.3: Execution**

```python
# 1. Build Circuit
qc = QuantumCircuit(n_qubits)
qc.h(range(n_qubits))

# 2. Grover Operator
oracle = bht_oracle_refined()
grover_op = GroverOperator(oracle)

# 1 Iteration is optimal for M=2, N=8
qc.append(grover_op, range(n_qubits))

# 3. Measure
qc.measure_all()

# 4. Run
sim = AerSimulator()
t_qc = transpile(qc, sim)
result = sim.run(t_qc, shots=1024).result()
counts = result.get_counts()

print("\nSearch Results:")
plot_histogram(counts)
```

**Interpretation**:

  * You should see peaks at `000` (0) and `100` (4).
  * `0` is the item we already knew (classically stored).
  * `4` is the **new item** found by the quantum computer.
  * Since $f(0) = 0 \pmod 4$ and $f(4) = 4 \pmod 4 = 0$, we have found a collision\!
  * **Significance**: Even though we only checked a small classical subset, the quantum search allowed us to instantly find the matching partner from the rest of the database.

# 16\. Element Distinctness

### **1. Algorithm Profile**

  * **Name**: Ambainisâ€™s Quantum Walk Algorithm for Element Distinctness.
  * **Origin**: Discovered by **Andris Ambainis** in **2004**.
  * **Type**: Oracular Algorithm (Quantum Walk).
  * **Speedup Class**: **Polynomial**. It solves the problem in **$O(N^{2/3})$** queries, significantly beating the classical limit of $O(N)$ (or $O(N \log N)$ for sorting-based methods).

### **2. Introduction & Context**

  * **The "Elevator Pitch"**: Given a list of $N$ items, are they all unique, or are there duplicates? Classically, youâ€™d verify this by sorting the list ($N \log N$) or using a hash table ($N$). A naive Grover search doesn't help much here (searching pairs takes $O(N)$). Ambainis found a brilliant way to use a **Quantum Walk**â€”a quantum particle jumping between groups of itemsâ€”to find duplicates faster than checking every item.
  * **Real-World Use Cases**:
      * **Database Integrity**: Quickly checking for duplicate records in massive unsorted databases.
      * **Cryptanalysis**: It generalizes the "Collision Finding" problem to functions with arbitrary domains, aiding attacks on cryptographic hash functions.
      * **Geometry**: Solving problems like "closest pair of points" often reduces to element distinctness.
  * **Current Status**: A fundamental theoretical result. It established **Quantum Walks** as a primary tool for algorithm design, distinct from the Quantum Fourier Transform.

### **3. Deep Theoretical Dive ðŸ§ **

#### **The Problem Statement**

Given a function $f: \{1, \dots, N\} \to Y$, determine if there exist two distinct indices $i, j$ such that $f(i) = f(j)$.
If such a pair exists, output the indices.

#### **The Classical Approach**

  * **Sorting**: Sort the values $f(x)$. Check adjacent elements. Cost: $O(N \log N)$.
  * **Hashing**: Store values in a hash table. Check for collisions. Cost: $O(N)$.

#### **The Quantum Mechanism: Walk on the Johnson Graph**

Grover's algorithm searches a static list. This algorithm searches a **graph of subsets**.

1.  **The Graph**: We construct a **Johnson Graph** $J(N, K)$.
      * Each "vertex" in this graph is a subset of $K$ items from the full list of $N$.
      * Two vertices are connected if you can get from one to the other by swapping one item (remove one, add one).
2.  **The State**: The quantum state holds a superposition of these subsets. For each subset, we also store the values of the items in that subset.
3.  **The Check**: A vertex is "marked" if the subset it represents contains a duplicate (a pair $i, j$ with $f(i)=f(j)$).
4.  **The Walk**: The algorithm performs a quantum walk on this graph.
      * It diffuses across the graph, exploring different combinations of $K$ items.
      * It uses **Grover-like reflection** to amplify the amplitude of "marked" vertices (subsets with duplicates).
5.  **Complexity Balance**:
      * Building the initial subset takes $K$ queries.
      * The walk requires roughly $N/K$ steps to explore the space.
      * Each step of the walk takes $\sqrt{K}$ time (to search the subset for collision).
      * **Optimization**: Balancing these costs leads to optimal $K = N^{2/3}$.
      * **Total Complexity**: $O(N^{2/3})$.

### **4. Practical Advantage & Complexity**

  * **Quantum Complexity**: $O(N^{2/3})$.
  * **Classical Complexity**: $O(N)$.
  * **The Advantage**: The speedup is $N^{1/3}$. For $N=1,000,000$, the classical computer takes 1 million steps; the quantum computer takes $\approx 10,000$ steps.
  * **Caveats**: The algorithm requires a large quantum memory (QRAM) to store the subsets and check for collisions efficiently. The space complexity is also $O(N^{2/3})$.

### **5. Implementation & Code Walkthrough (Qiskit) ðŸ’»**

Simulating the full Ambainis Walk is too complex for a basic notebook (requires complex graph topologies).
Instead, we will implement a **Grover-based Distinctness Checker**. This demonstrates the core *logic* of the check: searching the space of **all pairs** $(i, j)$ to find a collision.

  * **Search Space**: All pairs $(i, j)$ where $0 \le i < j < N$.
  * **Oracle**: Marks state $|i, j\rangle$ if $f(i) == f(j)$.

*Note: The complexity of this naive Grover approach is $\sqrt{N^2} = N$, which is no better than classical. The Quantum Walk (theoretical above) is needed for the true $N^{2/3}$ speedup, but this code demonstrates the quantum verification step.*

**Problem**: List $L = [1, 5, 3, 5]$. (Duplicate '5' at indices 1 and 3).
**Goal**: Find indices $(1, 3)$.

#### **Step 5.1: Defining the "Pair Check" Oracle**

We need an oracle that takes two indices $i, j$, retrieves $L[i]$ and $L[j]$, and flips the phase if they are equal (and $i \neq j$).

```python
import numpy as np
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
from qiskit.circuit.library import GroverOperator

# Problem Instance: Duplicate '5' at indices 1 and 3
# List indices: 0, 1, 2, 3
DATA_LIST = [1, 5, 3, 5]
N_ITEMS = len(DATA_LIST)
n_index_qubits = 2 # Need 2 qubits to represent index 0-3

def pair_check_oracle(data):
    # We need two registers for indices i and j.
    # Total qubits = 2 * n_index_qubits.
    qc = QuantumCircuit(2 * n_index_qubits)
    
    # We will implement a Phase Oracle manually by constructing the diagonal.
    # State |i>|j> gets phase -1 if data[i] == data[j] and i != j.
    
    num_states = 2**(2 * n_index_qubits)
    diagonal = np.ones(num_states)
    
    for state in range(num_states):
        # Decode state index into i and j
        # Binary: [j_bits] [i_bits]
        
        # Convert to binary string, pad to full length
        bin_str = format(state, f'0{2*n_index_qubits}b')
        
        # Split bits (Qiskit order is usually reversed but for diagonal logic we define our map)
        # Let's treat first n bits as i, next n as j
        # Note: standard big-endian bit string logic
        
        bits_i = bin_str[0:n_index_qubits]
        bits_j = bin_str[n_index_qubits:]
        
        i = int(bits_i, 2)
        j = int(bits_j, 2)
        
        # Check conditions
        # 1. Valid indices (always true here as N=4 matches 2 qubits)
        # 2. i != j (We don't care about comparing an item to itself)
        # 3. data[i] == data[j]
        
        if i != j and i < len(data) and j < len(data):
            if data[i] == data[j]:
                diagonal[state] = -1 # Mark collision
                
    qc.diagonal(list(diagonal), range(2 * n_index_qubits))
    qc.name = "Oracle(Distinctness)"
    return qc
```

#### **Step 5.2: The Search Circuit**

```python
# 1. Setup
qc = QuantumCircuit(2 * n_index_qubits)

# 2. Initialization (Superposition of all pairs)
qc.h(range(2 * n_index_qubits))

# 3. Grover Operator
oracle = pair_check_oracle(DATA_LIST)
grover_op = GroverOperator(oracle)

# Calculate optimal iterations
# Search space size = 16 (4*4).
# Solutions: (1,3) and (3,1). So 2 solutions.
# Iterations ~ pi/4 * sqrt(16/2) ~ 2.2 -> Round to 2.
iterations = 2

for _ in range(iterations):
    qc.append(grover_op, range(2 * n_index_qubits))

# 4. Measure
qc.measure_all()

print("Element Distinctness Circuit (Pair Search):")
qc.draw(output='text')
```

#### **Step 5.3: Execution & Analysis**

```python
# Run Simulation
sim = AerSimulator()
t_qc = transpile(qc, sim)
result = sim.run(t_qc, shots=1024).result()
counts = result.get_counts()

print("\nSearch Results (Index pairs i, j):")
# We verify the peaks correspond to the duplicates
for bitstring, count in counts.items():
    if count > 100:
        # Decode
        i = int(bitstring[0:n_index_qubits], 2)
        j = int(bitstring[n_index_qubits:], 2)
        
        val_i = DATA_LIST[i]
        val_j = DATA_LIST[j]
        
        status = "âœ… DUPLICATE" if val_i == val_j and i != j else "âŒ False Positive"
        print(f"Indices ({i}, {j}) -> Values ({val_i}, {val_j}) | Count: {count} | {status}")

plot_histogram(counts)
```

**Interpretation**:

  * You should see high probabilities for pairs `(1, 3)` and `(3, 1)`.
  * This confirms that indices 1 and 3 hold the same value (`5`).
  * **Pedagogical Note**: While this Grover approach works, it scales as $O(N)$. The advanced $O(N^{2/3})$ algorithm achieves its speedup by *not* searching every pair from scratch, but by walking between similar subsets and reusing information.
